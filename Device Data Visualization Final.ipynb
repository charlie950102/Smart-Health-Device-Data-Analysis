# Part 1: Measurement-specific visualizations for each patient    
This was done at the beginning of the project as a means to better understand the dataset, as well as check whether there existed faulty measurements of value 0.  
A sample plot of Patient 02's measurements for Heart rate (BP) are saved in the folder Images - Part 1

```python
patient = "patient 02"    # selecting patient to visualize
measurement = "Heart rate - from BP"     # selecting measurement type
df = pd.DataFrame({'Day':[], 'Value':[]})
count = 0
unit = 'unit'
pdate = list(set(data[data.email == patient].date_measurement))
pdate.sort()
pdate


# Obtaining values/dates for some patient's measurement type
for i in range(0,8905):
    if data.email[i] == patient and data.type_measurement[i] == measurement:   # patient and measurement
        if data.value_measurement[i] != 0:
            df.loc[i, 'Day'] = data.date_measurement[i]
            df.loc[i,'Value'] = data.value_measurement[i]
            unit = data.unit_measurement[i]
        else:
            count += 1


df = df.sort_values(by = 'Day')
df.reset_index(inplace=True)
del df['index']

title = "{} Vital Signs: {}".format(string.capwords(patient), measurement)
plt.figure(figsize=(15,7))
plt.plot(df['Day'],df['Value'], 'bo-', markersize = 2, linewidth = 1)
plt.title(title, fontsize = 17)
plt.xlabel('Day', fontsize = 15)
plt.ylabel(string.capwords(unit), fontsize = 15)
plt.xticks(rotation=90)
print("Total of", size(df['Day']), "meaningful measurements for:", measurement)
print("Total of", count, "meaningless measurements")
plt.savefig(string.capwords(title) + '.png', bbox_inches = 'tight')
```

# Part 2: Patient Vital Signs  
In order to get an idea of the distribution of data for each patient, I plotted each patient's measurements for each measurement type. I also drew a vertical red line in each plot that signals the 30-day mark, in order to see whether subsequent analysis of custom risk scores should include data up to day 30, or extend further to day 60. From these plots, we concluded that not enough patients had substantial data past day 30, hence we decided to utilize patient measurements only from days 0-30 in subsequent analyses. In this section, I only display the code used to generate visualizations for patient 02.   
Visualizations for all patients are saved in the folder Images - Part 2

```python
# Measurement Analysis for a SINGLE PATIENT

#for z in range(0,len(p)):

patient = "patient 02"       # select patient
measurement = ["Heart rate - from BP", "Heart rate - from Scale", "Heart rate - from Steel HR - Average over the day",
               "Systolic blood pressure", "Diastolic blood pressure", "Weight", "Number of steps"]
df = pd.DataFrame({'Day':[], 'BP':[], 'Scale':[], 'Steel HR':[], 'Systolic':[], 'Diastolic':[], 'Kg':[], 'Steps':[]})
pdate = list(set(data[data.email == patient].date_measurement))
pdate.sort()
pdate

# Obtaining values/dates for some patient's measurement type
for i in range(0, len(pdate)):
    for j in range(0, 8905):
        if data.email[j] == patient and data.date_measurement[j] == pdate[i]:
            df.loc[i, 'Day'] = pdate[i]
            if data.type_measurement[j] == measurement[0] and data.value_measurement[j] != 0:
                df.loc[i, 'BP'] = data.value_measurement[j]
            if data.type_measurement[j] == measurement[1] and data.value_measurement[j] != 0:
                df.loc[i, 'Scale'] = data.value_measurement[j]
            if data.type_measurement[j] == measurement[2] and data.value_measurement[j] != 0:
                df.loc[i, 'Steel HR'] = data.value_measurement[j]
            if data.type_measurement[j] == measurement[3] and data.value_measurement[j] != 0:
                df.loc[i, 'Systolic'] = data.value_measurement[j]
            if data.type_measurement[j] == measurement[4] and data.value_measurement[j] != 0:
                df.loc[i, 'Diastolic'] = data.value_measurement[j]
            if data.type_measurement[j] == measurement[5] and data.value_measurement[j] != 0:
                df.loc[i, 'Kg'] = data.value_measurement[j]
            if data.type_measurement[j] == measurement[6] and data.value_measurement[j] != 0:
                df.loc[i, 'Steps'] = data.value_measurement[j]


df = df.sort_values(by = 'Day')
df.reset_index(inplace=True)
del df['index']

# Plotting
d = list(range(1,len(df.Day)+1))

fig, axes = plt.subplots(7, figsize=(10,25), sharex=True)    # size is x by y
axes[0].plot(d,df['BP'], 'bo-', markersize = 2, linewidth = 1)
axes[1].plot(d,df['Scale'], 'bo-', markersize = 2, linewidth = 1)
axes[2].plot(d,df['Steel HR'], 'bo-', markersize = 2, linewidth = 1)
axes[3].plot(d,df['Systolic'], 'bo-', markersize = 2, linewidth = 1)
axes[4].plot(d,df['Diastolic'], 'bo-', markersize = 2, linewidth = 1)
axes[5].plot(d,df['Kg'], 'bo-', markersize = 2, linewidth = 1)
axes[6].plot(d,df['Steps'], 'bo-', markersize = 2, linewidth = 1)

# Set axes labels
plt.xlabel('Day', fontsize = 8)
plt.xticks(np.arange(min(d), max(d)+1, 1.0), fontsize = 8)
axes[0].set_ylabel('Bpm', fontsize = 8)
axes[1].set_ylabel('Bpm', fontsize = 8)
axes[2].set_ylabel('Bpm', fontsize = 8)
axes[3].set_ylabel('mmHg', fontsize = 8)
axes[4].set_ylabel('mmHg', fontsize = 8)
axes[5].set_ylabel('Kg', fontsize = 8)
axes[6].set_ylabel('Steps', fontsize = 8)
plt.tick_params(axis='y', labelsize=8)

# Set axes limits
axes[0].set_ylim([35, 144])
axes[1].set_ylim([35, 144])
axes[2].set_ylim([35, 144])
axes[3].set_ylim([68, 187])
axes[4].set_ylim([46, 137])
axes[5].set_ylim([5.1, 176])
axes[6].set_ylim([min(df.Steps), max(df.Steps)])

# Set titles
axes[0].set_title('Heart rate - from BP', fontsize = 10)
axes[1].set_title('Heart rate - from Scale', fontsize = 10)
axes[2].set_title('Heart rate - from Steel HR - Average over the day', fontsize = 10)
axes[3].set_title('Systolic BP', fontsize = 10)
axes[4].set_title('Diastolic BP', fontsize = 10)
axes[5].set_title('Weight', fontsize = 10)
axes[6].set_title('Number of steps', fontsize = 10)

# Plotting vertical line at day 30
for i in range(0,len(measurement)):
    if len(d) >= 30:
        axes[i].axvline(x = d[29], color = 'red')

plt.tight_layout()

title = "{} Vital Signs".format(patient)
plt.suptitle(string.capwords(title), x = 0.524, y = 1.01)

#plt.savefig(string.capwords(title) + '.png', bbox_inches = 'tight')
```

# Part 3: Number of patients with data per day  
To further get a better understanding of the data, I created histograms for each measurement type that show the data volume day by day. From these plots, we saw that other than for 'Heart Rate from Scale', each measurement type saw a steady decrease in patient data input as the study went on. This further helped us decide to use only measurements from day 0-30 for subsequent analyses.  
Visualizations are saved in the folder Images - Part 3.


```python
# Empty dataframe for days
dh = pd.DataFrame(index = range(0, 100), columns = [])

# Empty dataframe for patients using iteration
e = list(set(data['email']))
e.sort()

measurementH = 'Number of steps'

pdata = pd.DataFrame({e[0]: [], e[1]: [], e[2]: [], e[3]: [], e[4]: [], e[5]: [], e[6]: [], e[7]: [], e[8]: [], e[9]: [], 
                     e[10]: [], e[11]: [], e[12]: [], e[13]: [], e[14]: [], e[15]: [], e[16]: [], e[17]: [], e[18]: [], 
                      e[19]: [], e[20]: [], e[21]: [], e[22]: [], e[23]: [], e[24]: [], e[25]: [], e[26]: [], e[27]: [],
                     e[28]: [], e[29]: [], e[30]: [], e[31]: [], e[32]: [], e[33]: [], e[34]: [], e[35]: [], e[36]: []})

for i in range(0,len(e)):    # filling out pdata
    c = data[data.email == e[i]]
    d = c[c.type_measurement == measurementH]     # select measurement type
    d = d.sort_values('date_measurement')
    d = d.drop_duplicates()    # dropping duplicate rows in dataframe
    d.reset_index(inplace=True)
    del d['index']

    if not d.date_measurement.empty:
        start = datetime.strptime(min(d.date_measurement), '%Y-%m-%d')
    else:
        start = datetime.strptime('2017-06-28', '%Y-%m-%d')
    #end = datetime.strptime(max(d.date_measurement), '%Y-%m-%d')
    #step = datetime.timedelta(days=1)
    dlist = pd.date_range(start, periods=100).to_pydatetime().tolist() # day 1-100

    for j in range(0, len(dlist)):
        for z in range(0, len(d)):
            if dlist[j] == datetime.strptime(d.date_measurement[z], '%Y-%m-%d'): ##
                pdata.loc[j, e[i]] = 1

dh = dh.join(pdata)
dh['Count'] = dh.sum(axis=1)
dh.head()
```

```python
# Plotting

plt.figure(figsize=(30,7))
plt.bar(dh.index, dh['Count'])

title = "Number of Patients with {} Data per Day".format(measurementH)
plt.title(title, fontsize = 20)
#plt.xticks(rotation=90)
plt.xlabel('Day', fontsize = 17)
plt.ylabel('Data Volume', fontsize = 17)
#plt.savefig(string.capwords(title) + '.png', bbox_inches = 'tight')
```

# Part 4: Number of data for days     
I also created a plot for each measurement type that shows how the number of measurements each patient has per day changes over time. From these plots, we found out that the maximum number of measurements per day any given patient had was 3, specifically for the 4 measurement types: Systolic blood pressure, Diastolic blood pressure, Heart Rate (from BP), and Weight.
Based on these results, we decided to focus on these 4 measurement types in subsequent analyses.  
Visualizations are saved in the folder Images - Part 4.

```python
# 4a) Quantifying number of patient data for days

# Empty dataframe for days
dh = pd.DataFrame(index = range(0, 180), columns = [])

# Empty dataframe for patients using iteration
e = list(set(data['email']))
e.sort()

measurementH = 'Diastolic blood pressure'

pdata = pd.DataFrame({e[0]: [], e[1]: [], e[2]: [], e[3]: [], e[4]: [], e[5]: [], e[6]: [], e[7]: [], e[8]: [], e[9]: [], 
                     e[10]: [], e[11]: [], e[12]: [], e[13]: [], e[14]: [], e[15]: [], e[16]: [], e[17]: [], e[18]: [], 
                      e[19]: [], e[20]: [], e[21]: [], e[22]: [], e[23]: [], e[24]: [], e[25]: [], e[26]: [], e[27]: [],
                     e[28]: [], e[29]: [], e[30]: [], e[31]: [], e[32]: [], e[33]: [], e[34]: [], e[35]: [], e[36]: []})

for i in range(0,len(e)):    # filling out pdata
    c = data[data.email == e[i]]
    d = c[c.type_measurement == measurementH]     # select measurement type
    d = d.sort_values('date_measurement')
    #d = d.drop_duplicates()    # dropping duplicate rows in dataframe
    d.reset_index(inplace=True)
    del d['index']
    dlistunique = list(set(d.date_measurement))
    dlistunique.sort()

    if not d.date_measurement.empty:
        start = datetime.strptime(min(d.date_measurement), '%Y-%m-%d')
    else:
        start = datetime.strptime('2017-06-28', '%Y-%m-%d')
    #end = datetime.strptime(max(d.date_measurement), '%Y-%m-%d')
    #step = datetime.timedelta(days=1)
    dlist = pd.date_range(start, periods=180).to_pydatetime().tolist() # day 1-100 by date

    for j in range(0, len(dlist)):
        for z in range(0, len(d)):
            if dlist[j] == datetime.strptime(d.date_measurement[z], '%Y-%m-%d'): ##
                pdata.loc[j, e[i]] = len(d[d.date_measurement == d.date_measurement[z]])

dh = dh.join(pdata)
```

```python
# 4b) Plotting

#dh.loc[[0]] # row 0
#dh.loc[0:1] # row 0 to 1
#dh.loc[0,'patient 03']
#dh.iat[0,0]  # integer index

# 0 data points for that day
dh['Line 1'] = 0
for i in range(0, len(dh)):
    count = 0
    for j in range(0, 36):
        if np.isnan(dh.iat[i,j]) == True:
            count = count + 1
    dh.loc[i, 'Line 1'] = count

# 1 data point for that day
dh['Line 2'] = 0
for i in range(0, len(dh)):
    count = 0
    for j in range(0, 36):
        if dh.iat[i,j] == 1:
            count = count + 1
    dh.loc[i, 'Line 2'] = count

# 2 data points for that day
dh['Line 3'] = 0
for i in range(0, len(dh)):
    count = 0
    for j in range(0, 36):
        if dh.iat[i,j] == 2:
            count = count + 1
    dh.loc[i, 'Line 3'] = count

# 3 data points for that day
dh['Line 4'] = 0
for i in range(0, len(dh)):
    count = 0
    for j in range(0, 36):
        if dh.iat[i,j] == 3:
            count = count + 1
    dh.loc[i, 'Line 4'] = count

# 4 or more data points for that day
dh['Line 5'] = 0
for i in range(0, len(dh)):
    count = 0
    for j in range(0, 36):
        if dh.iat[i,j] >= 4:
            count = count + 1
    dh.loc[i, 'Line 5'] = count
#dh


# Plotting
plt.figure(figsize=(30,7))
xaxis = list(range(1, 181))
title = "Number of {} Data for Days".format(measurementH)
plt.title(title, fontsize=16)
plt.xlabel("Day")
plt.ylabel("Data Volume")
plt.plot(xaxis, dh['Line 1'], label = '0 data points')
plt.plot(xaxis, dh['Line 2'], label = '1 data points')
plt.plot(xaxis, dh['Line 3'], label = '2 data points')
plt.plot(xaxis, dh['Line 4'], label = '3 data points')
plt.plot(xaxis, dh['Line 5'], label = '4 or more data points')
plt.legend(loc='best')
#plt.savefig(title + '.png', bbox_inches = 'tight')
```
